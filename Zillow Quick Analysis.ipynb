```python
houseData = pd.read_csv("Neighborhood_Zhvi_SingleFamilyResidence.csv", encoding = "latin1")
houseData.head()
```

```python
data = houseData.fillna(method = "bfill", limit = 100, axis = 1)
actualData = data.copy(deep=True)
data
```

```python
# Dropping unecessary column
data = data.drop(["CountyName","SizeRank","Metro"], axis=1)
```

```python
# Finding the changes in the last 10 years
data["Rate of change(10 Years)"] = ((data['2019-11']-data['2009-11'])/data['2019-11'])

# Finding the changes in the last 5 years
data["Rate of change(5 Years)"] = ((data['2019-11']-data['2014-11'])/data['2014-11'])

# Finding the changes in the last 3 years
data["Rate of change(3 Years)"] = ((data['2019-11']-data['2016-11'])/data['2016-11'])

data.head()
```

```python
data = data.drop('RegionID',axis=1)
```

```python
data.groupby('State').sum().sort_values(by="Rate of change(10 Years)", ascending=False)["Rate of change(10 Years)"].plot(kind="bar", figsize=(20,5))
plt.show()
```

```python
data.groupby('State').sum().sort_values(by="Rate of change(5 Years)", ascending=False)["Rate of change(5 Years)"].plot(kind="bar", figsize=(20,5))
plt.show()
```

```python
data.groupby('State').sum().sort_values(by="Rate of change(3 Years)", ascending=False)["Rate of change(3 Years)"].plot(kind="bar", figsize=(20,5))
plt.show()
```

```python
data.groupby('State').sum().sort_values(by="Rate of change(3 Years)",ascending=False)[["Rate of change(10 Years)","Rate of change(5 Years)","Rate of change(3 Years)"]].plot(kind="bar", figsize=(20,5))
plt.show()
```
